### Pull the date, songs/artist and urls for each song from the html for each weekly chart's webpage

```python
dates=[]
rows = []
urls=[]
base_url= "https://musicchartsarchive.com"
week_directory = "/Users/karinashedrofsky/LEDE_2023/lyrics-project/weekly-htmls"
for file_name in os.listdir(week_directory):
    if file_name.endswith(".html"):  # Filter HTML files
        file_path = os.path.join(week_directory, file_name)
        
        # Open the HTML file
        with open(file_path, "r", encoding="utf-8") as file:
            html_content = file.read()
        
        # Create a BeautifulSoup object
        week_soup = BeautifulSoup(html_content)
        
        # Extract the title from the HTML
        title = week_soup.select("header h1#page-title")[0].text.strip()
        
        # Extract the date from the title using using regex
        date_match = re.search(r'\d{4}-\d{2}-\d{2}', title)
        if date_match:
            date = date_match.group()
            dates.append(date)
            
        #Extract each row from the table on each page
        row_els = week_soup.select("table tr")
        for row in row_els:
            cells = []
            for cell in row.select("td"):
                cells.append(cell.text.strip())
            rows.append(cells[:-1])

        #Extract the list of urls from each song on each page to use to scrape the lyrics and print NA if there's no url
        for row in row_els:
            td_tags = row.select("td")  #select all td tags within each row
            if len(td_tags) >= 2:
                second_td = td_tags[1]
                a_tag = second_td.find("a") #interested in the 2nd td tag because that's where song link would live
                if a_tag: 
                    href = a_tag.get("href")
                    urls.append(base_url + href) #pull the link and append it to the list if there's an <a> tag
                else:
                    urls.append("NA")
            else:
                urls.append("NA")
```

### Scrape the song name, artist name and lyrics from each song's html

```python
titles= []
artists= []
lyrics= []
song_directory = "/Users/karinashedrofsky/LEDE_2023/lyrics-project/song-htmls" 
for file_name in os.listdir(song_directory):
    if file_name.endswith(".html"):  # Filter HTML files
        file_path = os.path.join(song_directory, file_name)
        
        # Open the HTML file
        with open(file_path, "r", encoding="utf-8") as file:
            html_content = file.read()
        
        # Create a BeautifulSoup object
        song_soup = BeautifulSoup(html_content)
        
        #Pull the title from each song page
        title = song_soup.select("h1#page-title")[0].text.strip()
        titles.append(title)
        
        #Pull the artist from each song page
        artist = song_soup.select("h2 div.field-item")[0].text.strip()
        artists.append(artist)
        
        #Pull the lyrics from each song page
        elements = song_soup.select("div.field-items div.even") #pull all elements in 'div.even'
        lyric = sorted(elements, key=lambda element: len(element.text))[-1].text #take the longest one from each page to get the lyrics
        lyrics.append(lyric)
```

### Count how often each slang word appears in each song

```python
# Target word to count
target_word = "dope"

# Pattern with word boundary
pattern = r"\b" + re.escape(target_word) + r"\b"

dope_count_per_song = []
for lyric in lyrics:
    count = len(re.findall(pattern, lyric.lower()))
    dope_count_per_song.append(count)

sorted(dope_count_per_song, reverse=True)
```
